```python
def water_jug_memo(cap1, cap2, target):
    memo = set()
    path = []

    def dfs(j1, j2):
        if j1 == target or j2 == target:
            path.append((j1, j2))
            return True

        if (j1, j2) in memo:
            return False

        memo.add((j1, j2))
        path.append((j1, j2))

        # All possible moves
        moves = [
            (cap1, j2),   # fill jug1
            (j1, cap2),   # fill jug2
            (0, j2),      # empty jug1
            (j1, 0),      # empty jug2
            (j1 - min(j1, cap2 - j2), j2 + min(j1, cap2 - j2)),  # jug1 -> jug2
            (j1 + min(j2, cap1 - j1), j2 - min(j2, cap1 - j1))   # jug2 -> jug1
        ]

        for next_state in moves:
            if dfs(next_state[0], next_state[1]):
                return True

        path.pop()
        return False

    if dfs(0, 0):
        return path
    else:
        return None


# -----------------------------
# Take input from user
# -----------------------------
cap1 = int(input("Enter capacity of Jug 1: "))
cap2 = int(input("Enter capacity of Jug 2: "))
target = int(input("Enter target amount: "))

result = water_jug_memo(cap1, cap2, target)

if result:
    print("\nSolution path:")
    for step in result:
        print(step)
else:
    print("\nNo solution possible.")
```
